### 3.5 Traitement des dates
Conversion de la colonne `Date_Pub` en format datetime et extraction des composantes temporelles.

**Colonnes cr√©√©es** :
- `Annee` : ann√©e de publication
- `Mois` : mois de publication
- `Jour` : jour de publication

> **Note** : L'extraction du jour et du mois peut √™tre utile pour analyser la saisonnalit√© des sorties.

```python
# Conversion de la colonne 'Date_Pub' en format date
df ['Date_Pub'] = pd.to_datetime (df ['Date_Pub'])

# Extraction de l'ann√©e, du mois et du jour de publication
df ['Annee'] = df ['Date_Pub'].dt.year
df ['Mois'] = df ['Date_Pub'].dt.month
df ['Jour'] = df ['Date_Pub'].dt.day
# Doute sur la pertinence de cette extraction (day & month, utile ?)


print("üìÖ Dates converties et extraites :")
print("-"*65)

print(df[['Anime', 'Date_Pub', 'Annee', 'Mois', 'Jour']].head(10))
print (f"\n‚úÖ Types de 'Date_Pub' apr√®s conversion : {df['Date_Pub'].dtype}")

print(f"\n‚úÖ Dates converties et extraites avec succ√®s.")
print("-"*65)
```

## 4. üßπ Nettoyage des donn√©es {#nettoyage}

### 4.1 Suppression des doublons
√âlimination des lignes dupliqu√©es et r√©initialisation de l'index pour maintenir une s√©quence continue.

**M√©thode** : `drop_duplicates()` + `reset_index(drop=True)`

```python
# Suppresion des doublons et reset de l'index pour √©viter les trous
df = df.drop_duplicates().reset_index(drop=True)

# Verification du nombre de lignes apr√®s suppression des doublons
print(f"‚úÖ Doublons supprim√©s. Nouveau nombre de lignes : {df.shape[0]}")
```

### 4.2 Traitement des valeurs manquantes
Remplacement de toutes les valeurs manquantes par la cha√Æne `"Non renseign√©"`.

**Strat√©gie** : Imputation par valeur constante pour √©viter la perte d'informations.

> **Alternative possible** : Selon les colonnes, on pourrait utiliser la moyenne, la m√©diane ou supprimer les lignes.

```python
# Remplacement des valeurs manquantes par "Non renseign√©"
df = df.fillna("Non renseign√©")

print("Nombre de valeurs manquantes apr√®s imputation :")
print(df.isnull().sum()) # V√©rification qu'il n'y a plus de valeurs manquantes
```

### 4.3 Nettoyage des colonnes textuelles
Normalisation des colonnes de type texte pour assurer la coh√©rence des donn√©es.

**Op√©rations effectu√©es** :
- Suppression des espaces en d√©but/fin (`str.strip()`)
- Mise en majuscule de la premi√®re lettre de chaque mot (`str.title()`)

**Colonnes concern√©es** : Anime, Genre_Tags, Source, Studio, Status

```python
# V√©rification des colonnes disponibles
print("üìã Colonnes disponibles dans le DataFrame :")
print(df.columns.tolist())
print("-" * 70)

# Colonnes textuelles √† nettoyer (v√©rifi√©es et adapt√©es au dataset r√©el)
cols_textuelles = ['Anime', 'Genre_Tags', 'Source', 'Studio', 'Status']

# Filtrer pour ne garder que les colonnes pr√©sentes dans le DataFrame
cols_existantes = [col for col in cols_textuelles if col in df.columns]

print(f"‚úÖ Colonnes √† nettoyer : {cols_existantes}")
print("-" * 70)

# Boucle pour nettoyer les espaces et mettre en majuscule la premi√®re lettre de chaque mot
for col in cols_existantes:
    # V√©rifier que la colonne est bien de type texte
    if df[col].dtype == 'object':
        df[col] = df[col].astype(str).str.strip().str.title()

print(f"‚úÖ {len(cols_existantes)} colonnes textuelles nettoy√©es avec succ√®s")
```

## 5. ‚úÖ V√©rification finale de la qualit√© des donn√©es {#verification}

Contr√¥le exhaustif pour s'assurer que le dataset est pr√™t pour l'analyse et la sauvegarde.

**Points v√©rifi√©s** :
- Absence de valeurs manquantes
- Absence de doublons
- Cardinalit√© des colonnes
- Statistiques descriptives
- Coh√©rence des types de donn√©es
- D√©tection de valeurs aberrantes (scores n√©gatifs, ann√©es impossibles)
- Espaces parasites dans les colonnes textuelles

```python
# === V√âRIFICATION FINALE DE LA QUALIT√â DES DONN√âES ===
print("="*70)
print("üîç V√âRIFICATION FINALE DE LA QUALIT√â DES DONN√âES")
print("="*70)

# 1. Valeurs manquantes
print("\n1Ô∏è‚É£ Valeurs manquantes :")
print(f"   Total : {df.isnull().sum().sum()}")

# 2. Doublons
print("\n2Ô∏è‚É£ Doublons :")
print(f"   Total : {df.duplicated().sum()}")

# 3. Valeurs uniques par colonne (pour d√©tecter des anomalies)
print("\n3Ô∏è‚É£ Cardinalit√© des colonnes :")
print("-"*70)
for col in df.columns:
    n_unique = df[col].nunique()
    print(f"   {col:20} : {n_unique:6} valeurs uniques")

# 4. Statistiques sur colonnes num√©riques
print("\n4Ô∏è‚É£ Statistiques descriptives (colonnes num√©riques) :")
print("-"*70)
print(df.describe())

# 5. V√©rification des types de donn√©es
print("\n5Ô∏è‚É£ Types de donn√©es :")
print("-"*70)
print(df.dtypes)

# 6. V√©rification de valeurs aberrantes potentielles
print("\n6Ô∏è‚É£ Valeurs aberrantes potentielles :")
print("-"*70)

# Score n√©gatif ?
if 'Score' in df.columns:
    scores_negatifs = (df['Score'] < 0).sum()
    print(f"   Scores n√©gatifs : {scores_negatifs}")
    
# Ann√©es aberrantes ?
if 'Annee' in df.columns:
    annee_min = df['Annee'].min()
    annee_max = df['Annee'].max()
    print(f"   Plage d'ann√©es : {annee_min} - {annee_max}")
    if annee_min < 1900 or annee_max > 2026:
        print("   ‚ö†Ô∏è  Ann√©es suspectes d√©tect√©es !")

# 7. Espaces en d√©but/fin de cha√Ænes (colonnes texte)
print("\n7Ô∏è‚É£ Espaces parasites (colonnes texte) :")
print("-"*70)
text_cols = df.select_dtypes(include=['object']).columns
espaces_detectes = False
for col in text_cols:
    has_spaces = (df[col].astype(str).str.strip() != df[col].astype(str)).sum()
    if has_spaces > 0:
        print(f"   ‚ö†Ô∏è  {col} : {has_spaces} valeurs avec espaces parasites")
        espaces_detectes = True

if not espaces_detectes:
    print("   ‚úÖ Aucun espace parasite d√©tect√©")

print("\n" + "="*70)
print("‚úÖ V√©rification termin√©e !")
print("="*70)
```

### 7.1 Pr√©paration de l'environnement de visualisation

Importation des biblioth√®ques de visualisation et initialisation du dataset pour l'analyse.

**Op√©rations effectu√©es** :
1. Configuration du th√®me graphique (`whitegrid`)
2. Rechargement du dataset nettoy√© depuis `animes_cleaned.csv`
3. **Filtrage des donn√©es** : suppression des animes avec `Note_Pire_Ep = 0` (donn√©es incompl√®tes)

> **‚ö†Ô∏è Important** : Les animes avec une note de pire √©pisode = 0.0 sont exclus car ils correspondent g√©n√©ralement √† des s√©ries trop r√©centes ou sans donn√©es compl√®tes.

### 7.3 Distribution des notes par type d'√©pisode

Visualisation graphique de la r√©partition des notes pour comprendre la qualit√© globale des animes.

#### üìä Graphique 1 : Histogrammes comparatifs

**Type de graphique** : Histogramme avec courbe de densit√© (KDE - Kernel Density Estimation)

**Trois distributions affich√©es** :
1. **Pire √âpisode** (bleu) : montre la variabilit√© de la qualit√© minimale
2. **Meilleur √âpisode** (saumon) : indique les pics de qualit√© atteints
3. **Note Globale** (vert) : synth√®se √©quilibr√©e entre les extr√™mes

**Param√®tres** :
- `bins=30` : 30 intervalles pour une granularit√© fine
- `kde=True` : courbe liss√©e pour visualiser la tendance
- `figsize=(18,6)` : format large pour comparaison c√¥te √† c√¥te

**Interpr√©tation attendue** :
- Les **meilleurs √©pisodes** devraient √™tre concentr√©s vers les notes √©lev√©es (8-10)
- Les **pires √©pisodes** montrent la qualit√© minimale acceptable (souvent 5-7)
- La **note globale** se situe entre les deux extr√™mes

#### üìä Graphique 2 : Nuage de points synth√©tique (Strip Plot)

**Type de graphique** : Strip plot (nuage de points avec dispersion)

**Transformation des donn√©es** :
- Utilisation de `melt()` pour restructurer les donn√©es en format "long"
- Regroupement des 3 types de notes dans une seule colonne `Type_Note`
- Permet la comparaison visuelle directe des distributions

**Param√®tres** :
- `jitter=True` : dispersion horizontale pour √©viter le chevauchement des points
- `alpha=0.8` : transparence l√©g√®re pour voir les superpositions
- `palette='viridis'` : palette de couleurs moderne et accessible

**Avantages de cette visualisation** :
- Vue d'ensemble compacte des 3 distributions
- D√©tection rapide des valeurs aberrantes (outliers)
- Comparaison intuitive des m√©dianes et √©tendues

> **üí° Astuce** : Les zones denses du nuage indiquent les plages de notes les plus fr√©quentes. Une dispersion importante r√©v√®le une grande variabilit√© de qualit√© entre les √©pisodes.

```python
# Visualisation de la distribution des notes des pires √©pisodes
fig, axes = plt.subplots(1, 3, figsize=(18,6))

# Graphique 1 : Distribution des notes des pires √©pisodes
sns.histplot(df['Note_Pire_Ep'], bins=30, kde=True, ax=axes[0], color='skyblue')
axes[0].set_title('Pire √âpisode - La vraie diff√©rence')

# Graphique 2 : Distribution des notes des meilleurs √©pisodes
sns.histplot(df['Note_Meilleur_Ep'], bins=30, kde=True, ax=axes[1], color='salmon')
axes[1].set_title('Meilleur √âpisode - Tout les animes sont bon')

# Graphique 3 : Distribution des notes des meilleurs √©pisodes
sns.histplot(df['Note_Globale'], bins=30, kde=True, ax=axes[2], color='green')
axes[2].set_title('Note Globale - Vue d\'ensemble')

plt.tight_layout()
plt.show()
```

```python
# Synth√®se des notes (Nuage de points)
plt.figure(figsize=(10,8))

# Transformation du donn√©es pour affichage
df_melted = df.melt(value_vars=['Note_Pire_Ep', 'Note_Meilleur_Ep', 'Note_Globale'],
                    var_name='Type_Note', value_name='Note')
# Affichage du nuage de points
sns.stripplot(data=df_melted, x='Type_Note', y='Note', jitter=True, alpha=0.8, palette='viridis')

plt.title('Synth√®se des Notes par Type d\'√âpisode')
plt.xlabel('Notes sur 10')
plt.grid(axis='y', linestyle='--', alpha=0.3)

plt.show()
```
